```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('C:\\Users\\PC\\Downloads\\Telegram Desktop\\newhousing2.csv')
```

```python
data
```

```python
data.drop(['housing_median_age','total_rooms','total_bedrooms','population','households','median_house_value','ocean_proximity'],axis='columns',inplace=True)
data.head()
```

```python
kmeans = KMeans(n_clusters=3)
data["Cluster"] = kmeans.fit_predict(data)
data["Cluster"] = data["Cluster"].astype("category")

data.head()
```

```python
sns.relplot(
    x="longitude", y="latitude", hue="Cluster", data=data, height=10,)
```
